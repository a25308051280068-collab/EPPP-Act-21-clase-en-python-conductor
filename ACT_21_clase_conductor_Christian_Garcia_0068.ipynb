```python
class Conductor:
    # El método __init__ define los atributos (datos)
    def __init__(self, nombre, placas, edad):
        self.nombre = nombre
        self.placas = placas
        self.edad = edad

    # Definición de las Funciones (Métodos)
    def transportar(self):
        print(f"El conductor {self.nombre} está transportando pasajeros en el vehículo {self.placas}.")

    def planificacion_de_rutas(self):
        print(f"Calculando la mejor ruta para evitar el tráfico... (Edad del conductor: {self.edad} años).")

    def mantenimiento_de_vehiculo(self):
        print(f"El vehículo con placas {self.placas} ha entrado a revisión técnica.")

# --- Creación de Objetos (Instancias) ---

# Conductor 1 con sus datos de ejemplo
conductor1 = Conductor("Juan Pérez", "ABC-123", 35)

# Conductor 2 con sus datos de ejemplo
conductor2 = Conductor("María López", "XYZ-987", 28)

# --- Ejecución de las funciones ---

print("--- Actividades del Conductor 1 ---")
conductor1.transportar()
conductor1.planificacion_de_rutas()
conductor1.mantenimiento_de_vehiculo()

print("\n--- Actividades del Conductor 2 ---")
conductor2.mantenimiento_de_vehiculo()
conductor2.transportar()
conductor2.planificacion_de_rutas()
print("programa terminadp")
print("Progarama realizado por Christian Garcia NL = 21")
```
